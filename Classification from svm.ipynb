<h1>Classification using SVM</h1>

```python
import pandas as pd
import numpy as np
from numpy import array
from numpy import hstack
```

```python
df = pd.read_csv('SentimenAndStockfinal.csv')

df.columns
```

```python
in_seq1 = df['pos'].to_numpy()
#in_seq2 = df['Open'].to_numpy()
#in_seq3 = df['neg'].to_numpy()
# in_seq4 = df['Close'].to_numpy()
out_seq = df['UpDown'].to_numpy()
# # convert to [rows, columns] structure
in_seq1 = in_seq1.reshape((len(in_seq1), 1))
#in_seq2 = in_seq2.reshape((len(in_seq2), 1))
#in_seq3 = in_seq3.reshape((len(in_seq3), 1))
# in_seq4 = in_seq4.reshape((len(in_seq4), 1))
out_seq = out_seq.reshape((len(out_seq), 1))
# # horizontally stack columns
dataset = hstack((in_seq1,out_seq))
print(dataset)
```

```python
from sklearn.model_selection import train_test_split

X = np.array(dataset)

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X[:,:-1], X[:,-1], test_size=0.3,random_state=40)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
print(y_pred)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",metrics.precision_score(y_test, y_pred))
```

```python
for i in range(101):
    X_train, X_test, y_train, y_test = train_test_split(X[:,:-1], X[:,-1], test_size=0.3,random_state=i)
    
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
#Import svm model
    from sklearn import svm

#Create a svm Classifier
    clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
    clf.fit(X_train, y_train)

#Predict the response for test dataset
    y_pred = clf.predict(X_test)
#print(y_pred)

#Import scikit-learn metrics module for accuracy calculation
    from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Model Precision: what percentage of positive tuples are labeled as such?
    print("Precision:",metrics.precision_score(y_test, y_pred))
    print("Random State:",i)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
 
# making predictions on the testing set
y_pred = gnb.predict(X_test)
 
# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```
